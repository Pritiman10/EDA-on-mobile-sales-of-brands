### Calculating the cost price (Real price) of the phone after deducting discounts
### Also calculating sales in millions 

```python
df['Real_price']= df['sales_price']-df['discount_percent']*df['sales_price']
df['sales_in_millions']= (df['sales_price']*df['sales'])/1000000
df.head()
```

# Exploratory Data Analysis on the brands

```python
df.columns      #checking the columns for features
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

### We can see in the picture above that the top three brands by sales are REALME, APPLE & XIAOMI having more than 40 million worth of sales. 
### If we see the units sold then the top three brands are Realme, Xiaomi & Poco. Even though APPLE is second highest in sales, it sells the least units indicating higher end luxurios smartphone market capture.
### Samsung is at the bottom of the table, raking up 20 million in sales only.

```python
brand_sales = df.groupby('brand')['sales_in_millions'].sum().reset_index()
brand_units_sold = df.groupby('brand')['sales'].sum().reset_index()

# Set up the figure and axes for subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))

# Plotting Brand vs. Total Sales
sns.barplot(x='sales_in_millions', y='brand', data=brand_sales, ax=axes[0], orient='h')
axes[0].set_xlabel('Total Sales in millions')
axes[0].set_ylabel('Brand')
axes[0].set_title('Brand vs. Total Sales')

# Plotting Brand vs. Total Units Sold
sns.barplot(x='sales', y='brand', data=brand_units_sold, ax=axes[1], orient='h')
axes[1].set_xlabel('Total Units Sold')
axes[1].set_ylabel('Brand')
axes[1].set_title('Brand vs. Total Units Sold')

# Adjust the layout to prevent overlapping
plt.tight_layout()

# Show the plots
plt.show()
```

```python

brand_sales_pie = df.groupby('brand')['sales_in_millions'].sum().plot(kind='pie',y='brand',autopct='%1.1f%%', startangle=140)

# Add the title at the bottom of the plot using plt.text
plt.text(0.5, -0.1, 'Percentage of Sales by Mobile Brand', ha='center', transform=brand_sales_pie.transAxes)

# Show the pie chart
plt.show()
```

### If we see the cost price range of the brands in the below plot, it is clear that Xioami, Realme and Poco cater to the low income groups of the population(6k-25k) while apple sells phones in the range of (28k-80k) on the e commerce platform.

```python
df.plot.scatter(x='brand',y='Real_price',alpha=0.5)
```

# Lets check the mobile specifications(rom, ram, battery, processor, color, ratings, screen size) that the brands are providing for the customer


# Some pointers to note :
1. Apple has 3 own processors that are water, ceramic and iOS(second graph from 1st row in the figure above) and other brands sell the common android processors like mediatek, qualcomm, exynos and others
2. Apple does not sell very large screen sized phone, in this data other brands sell a wide variety of RAM options starting 2GB to 12GB phones
3. The android segment brands comprising of Xiaomi, Realme, Poco and Samsung offer a wider range of battery capacities, so the customer has more variety to select from in the lower end mobile phone segments
4. All the brands perform well overall as the mean rating from the customer is 4+ for all of them

```python
# Create a larger figure size
plt.figure(figsize=(16, 10))

# List of features to include in the pairplot
features = ['base_color', 'processor', 'screen_size', 'ROM', 'RAM', 'display_size', 'num_rear_camera', 'num_front_camera', 'battery_capacity', 'ratings', 'discount_percent']

# Create a pairplot with subplots
num_cols = 4  # Number of columns in the subplot grid
num_rows = len(features) // num_cols + 1

for i, feature in enumerate(features):
    plt.subplot(num_rows, num_cols, i + 1)
    #sns.scatterplot(x=feature, y='sales_in_millions', data=df, alpha=0.7, label='Real_price', color='blue')
    sns.scatterplot(x='brand', y=feature, data=df, alpha=0.7, color='green')
    plt.title(f'{feature} vs. Real_price & sales')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)
# Adjust subplot layout
plt.tight_layout()

# Show the plot
plt.show()
```

# Below heatmap visualises the dependency of one feature with other. 
1. The negetive correlation (-0.23) of sales(units) and sales_price indicate that the market is dominated by mid range smart phones. More number of people prefer a variety of higher specifications at a lower price.
2. The positive correlation (0.95) of sales and number_of_ratings of 4+ stars indicates that customers check for reviews before purchasing. 


```python
correlation_matrix = df[['ROM','RAM','display_size','num_rear_camera','battery_capacity','ratings','num_of_ratings','discount_percent','Real_price','sales']].corr()
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
```
